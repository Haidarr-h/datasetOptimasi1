**KELAS GRAPH**

```python
class Graph:

  def __init__(self, amount_vertices):
    self.edges = {} #sambungan antar kotanya
    self.vertices = set() #list kotanya
    self.amount_vertices = amount_vertices #banyak lokasi kunjungan atau merepresentasikan setiap titik yang ada (lokasi kunjungannya)

  def existEdge(self, source, destination):
    return (True if (source, destination) in self.edges else False)

  def addEdge(self, source, destination, cost = 0):
    if not self.existEdge(source, destination):
      self.edges[(source, destination)] = cost #menginputkan ke dalam dictionary edges, contoh isi : {(69, 70): 7
      self.vertices.add(source) #menginputkan ke dalam set (tidak menerima duplikasi), isinya menjadi {69, 70}
      self.vertices.add(destination)

  def showGraph(self):
    print('Showing Graph:')

    for edge in self.edges:
      #print('%d linked in %d with cost %d' % (edge[0], edge[1], self.edges[edge]))
      print("source:", edge[0], "\tdestination :", edge[1], "\tcost:", self.edges[edge])

  def getCostPath(self, path):
    total_cost = 0
    for i in range(len(path) - 1): #path di sini umumnya jalur yang diajukan. Maka jumlah jalur tentunya memiliki total sebanyak 236 jalur
        if (path[i], path[i+1]) in self.edges:
            total_cost += self.edges[(path[i], path[i+1])]
        else:
            return float('inf')  # Assign a high cost to paths with missing edges

    # Add cost for returning to the start if needed
    '''
    if (path[-1], path[0]) in self.edges:
        total_cost += self.edges[(path[-1], path[0])]
    else:
        total_cost += float('inf')  # Again, ensure there's a high cost for missing return edge
    '''
    return total_cost

  #menghasilkan output path random sebanyak populasi atau partikel yang diinginkan pada parameter
  def getRandomPaths(self, max_size):
    #inisialisasi variabel path randomnya dan list vertices membentuk list yang berisi semua kota yang ada (yang sudah di add menggunakan addEdge)
    random_paths, list_vertices = [], list(self.vertices)

    #mengambil kota random dari list kota yang ada
    initial_vertice = random.choice(list_vertices)

    if initial_vertice not in list_vertices:
      print('Error: Kota %d yang dicari pada penentuan random Path tidak ditemukan!' % initial_vertice)
      sys.exit(1)

    #menghilangkan kota tersebut
    list_vertices.remove(initial_vertice)
    #menambahkan kota tersebut pada index pertama
    list_vertices.insert(0, initial_vertice)

    #perulangan sebanyak jumlah partikelnya
    for i in range(max_size):
      #list temporary kecuali kota initial_vertice
      list_temp = list_vertices[1:]
      #peracakan list temp
      random.shuffle(list_temp)
      #insert initial vertice di depan
      list_temp.insert(0, initial_vertice)

      if list_temp not in random_paths:
        random_paths.append(list_temp)

    return random_paths
```

**KELAS PARTIKEL**

```python
# Pembuatan Kelas Partikel
class Particle:
  def __init__(self, solution, cost):
    self.solution = solution
    self.pbest = solution
    self.cost_current_solution = cost
    self.cost_pbest_solution = cost
    self.velocity = []

  #PBest partikel
  def setPBest(self, new_pbest):
    self.pbest = new_pbest

  def getPBest(self):
    return self.pbest

  #kecepatan partikel
  def setVelocity(self, new_velocity):
    self.velocity = new_velocity

  def getVelocity(self):
    return self.velocity

  #path solusi pada iterasi tertentu dari suatu aprtikel
  def setCurrentSolution(self, solution):
    self.solution = solution

  def getCurrentSolution(self):
    return self.solution

  #cost dari PBest partikel
  def setCostPBest(self, cost):
    self.cost_pbest_solution = cost

  def getCostPBest(self):
    return self.cost_pbest_solution

  #penentuan cost dari solusi pada iterasi tertentu dari suatu partikel
  def setCostCurrentSolution(self, cost):
    self.cost_current_solution = cost

  def getCostCurrentSolution(self):
    return self.cost_current_solution

  #reset velocity
  def clearVelocity(self):
    del self.velocity[:]
```

**CLASS PSO**

```python
class PSO:

	def __init__(self, graph, iterations, size_population, beta=1, alfa=1):
		self.graph = graph #Pemanggilan kelas graph sebagai atributnya
		self.iterations = iterations
		self.size_population = size_population
		self.particles = [] #list dari kelas partikel (populasinya)
		self.beta = beta # the probability that all swap operators in swap sequence (gbest - x(t-1))
		self.alfa = alfa # the probability that all swap operators in swap sequence (pbest - x(t-1))

		#pembuatan solusi acak sebanyak jumlah partikel yang ditentukan
		#contoh outputnya : [[9, 2, 3, 4], [3, 2, 9, 4]] (tentunya isinya sebanyak 236 kota yang dimaksud)
		solutions = self.graph.getRandomPaths(self.size_population)

		#periksa apakah variabel solutions terdapat isinya atau tidak
		if not solutions:
			print('Populasi tidak berisi, coba jalankan program kembali !')
			sys.exit(1)

		# creates the particles and initialization of swap sequences in all the particles
		#Pembuatan partikels (setiap masing2 partikel yang akan mencari posisi terbaik)
		for solution in solutions:
			#Pembuatan partikel baru
			particle = Particle(solution=solution, cost=graph.getCostPath(solution))
			# add the particle
			self.particles.append(particle)

		#===di sini sudah dalam keadaan memiliki 10 partikel (10 solusi acak)===
		#update populasi just in case (walaupun sebenarnya sama)
		#print(size_population)
		#print(len(self.particles))
		self.size_population = len(self.particles)


	# set gbest (best particle of the population)
	def setGBest(self, new_gbest):
		self.gbest = new_gbest

	# returns gbest (best particle of the population)
	def getGBest(self):
		return self.gbest

	# shows the info of the particles
	def showsParticles(self):
		print('Menunjukan Partikel Sebagai Solusi\n')
		for particle in self.particles: #self particles berisikan list sebanyak dengan jumlah populasi yang diinputkan
        # Format cost values to handle None or float('inf') gracefully
			cost_pbest = 'Infinity' if particle.getCostPBest() == float('inf') else particle.getCostPBest() #akan berisikan 'infinity' apabila getCost menunjukan inf
			cost_current = 'Infinity' if particle.getCostCurrentSolution() == float('inf') else particle.getCostCurrentSolution()

			print(f'pbest: {particle.getPBest()}\t|\tcost pbest: {cost_pbest}\t|\tcurrent solution: {particle.getCurrentSolution()}\t|\tcost current solution: {cost_current}')
		print('')

	def run(self):

		# for each time step (iteration)
		for t in range(self.iterations):
        # Filter particles to ensure we only consider those with valid costs
				#memfilter partikel di mana hanya partikel yang memmiliki cost saja yang masuk ke dalam variabel valid_particles
			valid_particles = [p for p in self.particles if p.getCostPBest() is not None]

			#sampai saat ini valid_particles berisi list mengenai ramdom path dan solusinya

			if not valid_particles:
				print("Tidak Ada Partikels Yang Valid, Periksa Kembali !")
				return

      # Update gbest with valid particles only
			#mencari perhitungan solusi terkecil secara global (gBest)
			self.gbest = min(valid_particles, key=lambda p: p.cost_pbest_solution)
			self.setGBest(self.gbest)
			#print(min(valid_particles, key=lambda p: p.cost_pbest_solution))

			#'''
			#untuk melihat setiap pBest dari setiap partikel
			for i, particle in enumerate (valid_particles):
				print('iterasi ke', t, 'partikel ke', i, particle.cost_pbest_solution)

			#'''

			#print('gbest:', self.gbest.getPBest())
			x = 0
			# for each particle in the swarm
			for particle in self.particles:

				particle.clearVelocity() # cleans the speed of the particle
				temp_velocity = []
				solution_gbest = copy.copy(self.gbest.getPBest()) #Mendapatkan path (solusi) dari gbest #copy2 sepertinya dont matter
				#print('solusigevst', solution_gbest)
				solution_pbest = particle.getPBest()[:] #mendapatkan path(solusi) dari pbest pada current partikel, penggunaan [:] berfungsi agar melakukan shallow copy (bukan mengcopy objeknya)
				solution_particle = particle.getCurrentSolution()[:] #mendapatkan path (solusi) current solusi dari current partikel

				# generates all swap operato	rs to calculate (pbest - x(t-1))
				for i in range(self.graph.amount_vertices): #236 kota
					if solution_particle[i] != solution_pbest[i]: #perbandingan antara partikel current iteration dengan pbestnya
						# generates swap operator
						swap_operator = (i, solution_pbest.index(solution_particle[i]), self.alfa)
						#membentuk tuple (data koleksi) dari nilai (iterasi, index kota pada pbest yang sesuai dengan nilai solusi sekarang pada index sesuai iterasi, alfa )
						#print('swap', swap_operator)

						# append swap operator in the list of velocity
						temp_velocity.append(swap_operator)
						#print('temp', temp_velocity)

						# makes the swap
						aux = solution_pbest[swap_operator[0]]
						solution_pbest[swap_operator[0]] = solution_pbest[swap_operator[1]]
						solution_pbest[swap_operator[1]] = aux
						'''
						if x == 0:
							x = x + 1
							print('swapOps', swap_operator)

				print('tmp', temp_velocity)
				'''
				# generates all swap operators to calculate (gbest - x(t-1))
				for i in range(self.graph.amount_vertices):
					if solution_particle[i] != solution_gbest[i]:
						# generates swap operator
						swap_operator = (i, solution_gbest.index(solution_particle[i]), self.beta)

						# append swap operator in the list of velocity
						temp_velocity.append(swap_operator)

						# makes the swap
						aux = solution_gbest[swap_operator[0]]
						solution_gbest[swap_operator[0]] = solution_gbest[swap_operator[1]]
						solution_gbest[swap_operator[1]] = aux


				# updates velocity
				particle.setVelocity(temp_velocity)

				# generates new solution for particle
				for swap_operator in temp_velocity:
					if random.random() <= swap_operator[2]:
						# makes the swap
						aux = solution_particle[swap_operator[0]]
						solution_particle[swap_operator[0]] = solution_particle[swap_operator[1]]
						solution_particle[swap_operator[1]] = aux

				# updates the current solution
				particle.setCurrentSolution(solution_particle)
				# gets cost of the current solution
				cost_current_solution = self.graph.getCostPath(solution_particle)
				# updates the cost of the current solution
				particle.setCostCurrentSolution(cost_current_solution)

				# checks if current solution is pbest solution
				if cost_current_solution < particle.getCostPBest():
					particle.setPBest(solution_particle)
					particle.setCostPBest(cost_current_solution)
```
